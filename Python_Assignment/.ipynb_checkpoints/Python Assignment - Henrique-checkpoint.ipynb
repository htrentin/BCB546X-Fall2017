```python
## 2 ##
## Codon translate function #

def translate_function(string_nucleotides): 
    mito_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"]
    n =3
    codon = [string_nucleotides[i:i+n] for i in range(0,len(string_nucleotides),n)]
    aa_seq_string=[]
    for w in range(0,len(codon)-1,1):
        aa_seq_string.append(mito_table.forward_table[codon[w]]) 
    return(str(aa_seq_string))
```

#This chunck of code is used to translate DNA to aminoacids.
#The function Codon.Table.unambiguous_dna_by_name, from Biopython, contains the unambiguous codons that are translated to specific aminoacids. The ["Vertebrate Mitochondrial"] specification says that we want to use the genetic code of the mitochondria of vertebrata.
#An objected with length 3 is created, in order to split the DNA sequence in 3 base pairs (codons), which are translated to a aminoacid.
#An object called codon is created, in which the results of a for loop will be placed. The for loop is using the functions range and length, along with the pre-determined object of length =3, to separada the codons.
#An object called called aa_seq_string is created, in which an additional function will write AAs based on the codons.
#The for loop is asking the program to delete the lost codon, which is a stop codon. The removal of the last codon is done by the "-1".
#The result of the codon translation will be placed into the aa_seq_string object, using the mito_table.forward_table and the append functions.
#The data is coerced to a string with the str function and the loop is terminated with the return command. 


##Source: http://biopython.org/DIST/docs/api/Bio.Data.CodonTable-module.html

```python
## 3 ##
# My alternative translate function

from Bio.Seq import Seq
def my_alt_function(string_nucleotides):
    aa_seq_string = string_nucleotides.translate(table="Vertebrate Mitochondrial", to_stop=True)
    return(str(aa_seq_string))
```

```python
#Biopython has a translate function, which translates DNA to aminoacids. The reference table for this function is NCBI's table, and in case we want to use a different table we need to specify it(e.g.,"Vertebrate Mitochondrial"). The argument to_stop=True makes the translation process to stop after a stop codon is found.
#Information about the translate function can be found at: http://biopython.org/DIST/docs/tutorial/Tutorial.html#htoc25
```

```python
## 4 ##
## My molecular weight calculator function

from Bio.SeqUtils.ProtParam import ProteinAnalysis
def compute_molecular_weight(aa_seq_string):
    analysed_seq = ProteinAnalysis(aa_seq_string)
    protein_weight = analysed_seq.molecular_weight()
    return(protein_weight)
```

```python
#As suggested, ProtParam tools, from the SeqUtils package, have some useful functions for analyzing protein sequences. ProteinAnalysis function checks if the protein sequence is a string, and molecular weight function is used to calculate the molecular weight of a protein. Aminoacid sequence is already is a string format, because the "str" command was used in the previous code.
```

```python
## 5 ##
#My GC content function calculator:

def GC_content_calculator(aa_seq_string):
    count = 0
    for base in aa_seq_string:
        if base is 'G':
                count += 1
        if base is 'C':
                count += 1
    return((count)/len(aa_seq_string))
```

```python
## 6 ## 
## Add two new columns to the penguin DataFrame: (1) molecular weight and (2) GC content.
## Set the value to 'NaN' to indicate that these cells are currently empty.

penguins_df_copy['Molecular_Weight'] = ['NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN']
penguins_df_copy['GC_Content'] = ['NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN','NaN']
penguins_df_copy

#I created the two new columns using brackets. Using this command, I had to specify the value of each cell.
```

```python
## 7 ##
## A for loop that translates each sequence in the data set, calculate its molecular weight, and obtain its GC content.

# I did not find a way to place the results of the for loop directly into the penguins_df_copy dataframe, so I created two new variables and later appendend them to the dataframe.
Molecular_Weight = [] 
GC_Content = [] 

#The for loop:
for k, v in cytb_seqs_copy.items():  #Recalling the dictionary per key and value.
    if k in cytb_seqs: # To determine that the for loop is based on each key.
        GC_values = GC_content_calculator(cytb_seqs[k]) #I am calling the function to compute GC content I created before
        GC_Content.append(GC_values) #Appending the GC content of each key to the empty variable created.
        AA_sequence = my_alt_function(cytb_seqs[k]) #I am using the function I created before to translate DNA sequences to AAs.
        Mol_weight = compute_molecular_weight(AA_sequence) #Calculating the molecular weight of the resulting AA sequences with the function determined on item 4.
        Molecular_Weight.append(Mol_weight) # Appending the molecular weight each AA sequence to the variable I created before, using its key as reference.
        
penguins_df_copy['GC_Content'] = GC_Content #Appending the GC content of each key inside the penguins_df_copy dataframe.
penguins_df_copy['Molecular_Weight'] = Molecular_Weight #Appending the Molecular weight of each key inside the penguins_df_copy dataframe.
penguins_df_copy #Giving a look at the dataframe, to check if it looks as expected

#Comment: I calculate the GC content of each DNA sequence. From you description, it seems that you want the GC content of each aa sequence, which I do not think it is what you mean.
```
